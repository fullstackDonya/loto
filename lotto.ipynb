```python
import pandas as pd

# Dictionnaire pour stocker les DataFrames
datasets = {}

# Liste des fichiers avec noms personnalisés pour les clés
file_paths = {
    #Historique Loto
    "loto": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/loto.csv",
    "nouveau_loto": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/nouveau_loto.csv",
    "loto2017": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/loto2017.csv",
    "loto_201902": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/loto_201902.csv",
    "loto_201911": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/loto_201911.csv",

    # Historique Grand loto
    "lotonoel2017": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/lotonoel2017.csv",
    "grandloto_201912": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/grandloto_201912.csv",

    # Historique super loto
    "superloto2017": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/superloto2017.csv",
    "superloto_201907": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/superloto_201907.csv",
    "sloto": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/sloto.csv",
    "nouveau_superloto": "/content/drive/MyDrive/Colab Notebooks/loto/dataset/nouveau_superloto.csv"
}

# Lecture de chaque fichier dans un DataFrame séparé
for name, path in file_paths.items():
    try:
        df = pd.read_csv(path, sep=';')  # Adapter le séparateur si nécessaire
        datasets[name] = df
        print(f"{name} loaded: {df.shape[0]} rows")
    except Exception as e:
        print(f"Error loading {name}: {e}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import itertools

def analyze_combinations_scatter(dataset_name, combination_size=2):
    """Analyzes the frequency of number combinations and creates a scatter plot."""
    df = datasets[dataset_name]

    # Get all combinations of numbers
    combinations = []
    for index, row in df.iterrows():
        boules = [row['boule_1'], row['boule_2'], row['boule_3'],
                  row['boule_4'], row['boule_5']]
        combinations.extend(list(itertools.combinations(boules, combination_size)))

    # Count the frequency of each combination
    combination_counts = Counter(combinations)

    # Extract data for scatter plot
    x_values = [comb[0] for comb in combination_counts.keys()]
    y_values = [comb[1] for comb in combination_counts.keys()]
    sizes = [count for count in combination_counts.values()]

    # Create the scatter plot
    plt.figure(figsize=(10, 8))
    plt.scatter(x_values, y_values, s=sizes, alpha=0.5)  # s=sizes makes points larger for more frequent combinations
    plt.title(f"Scatter Plot of Lottery Number Combinations ({dataset_name})")
    plt.xlabel("Number 1")
    plt.ylabel("Number 2")
    plt.colorbar(label="Frequency")  # Add a colorbar to show frequency
    plt.show()

# Example usage
analyze_combinations_scatter('loto', combination_size=2)
```

```python
!pip install mpl_toolkits
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from collections import Counter
import itertools
import numpy as np

def analyze_combinations_3d(dataset_name, combination_size=2):
    """Analyzes the frequency of number combinations and creates a 3D surface plot."""
    df = datasets[dataset_name]

    # Get all combinations of numbers
    combinations = []
    for index, row in df.iterrows():
        boules = [row['boule_1'], row['boule_2'], row['boule_3'],
                  row['boule_4'], row['boule_5']]
        combinations.extend(list(itertools.combinations(boules, combination_size)))

    # Count the frequency of each combination
    combination_counts = Counter(combinations)

    # Create data for 3D plot
    x_values = [comb[0] for comb in combination_counts.keys()]
    y_values = [comb[1] for comb in combination_counts.keys()]
    z_values = [count for count in combination_counts.values()]

    # Create the 3D surface plot
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Use a meshgrid for the surface
    X, Y = np.meshgrid(np.unique(x_values), np.unique(y_values))
    Z = np.zeros_like(X, dtype=int)

    # Fill Z with frequencies
    for i, x in enumerate(np.unique(x_values)):
        for j, y in enumerate(np.unique(y_values)):
            comb = (x, y)
            if comb in combination_counts:
                Z[j, i] = combination_counts[comb]

    ax.plot_surface(X, Y, Z, cmap='viridis')  # Use a colormap
    ax.set_title(f"3D Surface Plot of Lottery Number Combinations ({dataset_name})")
    ax.set_xlabel("Number 1")
    ax.set_ylabel("Number 2")
    ax.set_zlabel("Frequency")
    plt.show()

# Example usage
analyze_combinations_3d('loto', combination_size=2)
```

```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt

# ... (your previous code to read CSV files into the 'datasets' dictionary) ...

# Function to get most frequent numbers
def get_most_frequent_numbers(dataset_name, num_numbers=5):
    """Returns the 'num_numbers' most frequent lottery numbers
       from the specified dataset.
    """
    df = datasets[dataset_name]
    all_boules = pd.concat([df['boule_1'], df['boule_2'], df['boule_3'],
                            df['boule_4'], df['boule_5']], ignore_index=True)

    # Get frequency counts of each number
    frequency = all_boules.value_counts()

    # Return the top 'num_numbers' most frequent numbers
    return frequency.index[:num_numbers].tolist()

# Example usage:
dataset_name = 'loto_201911'  # Choose your dataset
suggested_numbers = get_most_frequent_numbers(dataset_name)
print(f"Suggested numbers for {dataset_name}: {suggested_numbers}")
```
